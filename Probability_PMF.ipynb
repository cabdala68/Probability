**UNIFORM DAĞILIM var**, çünkü;

Dağılım her değer için aynı olasılık veriyor.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
num_throws = 10000
outcomes = np.zeros(num_throws)
for i in range(num_throws):
    # bir zar havaya atalım
    outcome = np.random.choice(['1', '2', '3', '4', '5', '6'])
    outcomes[i] = outcome

val, cnt = np.unique(outcomes, return_counts=True)
prop = cnt / len(outcomes)

# Zarı 10000 kez havaya attık, şimdi grafik olarak gösterelim
plt.bar(val, prop)
plt.ylabel("Olasılık")
plt.xlabel("Sonuç")
plt.show()
plt.close()
```

Dağılımın Düzgün bir dağılıma sahip olduğunu ve her sonucun yaklaşık **1/6 (≈0.17)**. bir olasılığa sahip olduğunu görebiliriz.
